```python
# Dictionary for each node, with paths to connected nodes and weight information,
# respectively
location_dic = {
    "S": [["A", 8], ["B", 4], ["D", 6]],
    "A": [["B", 3], ["G1", 9]],
    "B": [["A", 5], ["C", 3]],
    "C": [["S", 4], ["F", 5], ["H", 9]],
    "D": [["C", 5], ["E", 3]],
    "E": [["G2", 4], ["S", 10]],
    "F": [["B", 8], ["G1", 8]],
    "G1": [],
    "G2": [],
    "H": [["D", 10], ["G2", 6]]
}

# Define a list of the heuristics for the nodes
H = {
    "S" : 6,
    "A" : 4,
    "B" : 5,
    "C" : 6,
    "D" : 4,
    "E" : 6,
    "F" : 2,
    "G1" : 0,
    "G2" : 0,
    "H" : 3
}

# Variable for start node
start_node = "S"

# Define the goal states
goal_nodes = {"G2", "G1"}

# Visited nodes collects visited nodes
visited_dic = {}

# Used to store the path
tracking_dic = {}

# Start the agenda with the start position information
agenda = [["S", 0]]

step = 1

print("The following will present nodes with their f-value in brackets. This is found from the equation f(n) = g(n) + h(n).")
print("The g value is the total cost of the path from the start node to the current node.")
print("The h value is the heuristic of the node, the estimated cost to find a goal node\n")

# Start of the while loop while will run while the agenda is not empty
while agenda != [[]]:
    
    # Visualisation of each step for finding the goal node
    print("Step " + str(step) + ":")
    print("Agenda---------------" + "[", end = " ")
    for node in agenda:
        print(str(node[0]) + "(" + str(node[1] + H[node[0]]) + "),", end = " ")
    print("]")
    print("Visited--------------" + "[", end = " ") 
    for node in visited_dic:
          print(str(node) +  "(" + str(visited_dic[node] + H[node]) + "),", end = " ")
    print("]")
    
    # Current will be the first node in agenda
    current = agenda[0]

    
    step += 1
    # Loop to find lowest total path cost and heuristic cost within the agenda
    for node in agenda:
        check_node = H[node[0]] + node[1]
        current_node = H[current[0]] + current[1]
        if check_node < current_node and H[current[0]] <= node[1] :
            current = node
        
    # Remove the current node
    agenda.remove(current)

    # Check if the current is in the visited states
    if current[0] in visited_dic:
        # If the node is in visited, check that current has lower cost than same node in visited
        if current[1] < visited_dic.get((current[0])):
            # If it has, replace the old value in visited with new
            visited_dic[current[0]] = current[0]
        else:
            print("[" + str(current[0]) + "(" + str(current[1] + H[current[0]]) + ")]" +
                  " will not be expanded as it is already in visited and does not have a lower f-value.")
            print("New Agenda-----------" + "[", end = " ")
            for node in agenda:
                    print(str(node[0]) + "(" + str(node[1] + H[node[0]]) + "),", end = " ")
            print("]\n")
    else:
        # If the node is not in visited states add it
        visited_dic[current[0]] = current[1]
        
        # Visualisation of node to be expanded
        print("Node to be expanded--" + "[ " + str(current[0]) + "(" + str(current[1] + H[current[0]]) + ")]")
        
        # Check if current is a goal node
        if current[0] in goal_nodes:
            print("Found a Goal Node!!!\n")
            break
        # If it is not the goal node, expand its nodes
        else:
            expanded = location_dic.get(current[0])
            
            for node in expanded:
                # Add the path cost to the expanded nodes weight
                node[1] = node[1] + current[1]
                # Add the node to tracking dictionary with cost
                listToStr = ''.join([str(elem) for elem in node])
                tracking_dic[listToStr] = current

                # Add the new nodes to the agenda
                agenda.append(node)
            
            print("New Agenda-----------" + "[", end = " ")
            for node in agenda:
                    print(str(node[0]) + "(" + str(node[1] + H[node[0]]) + "),", end = " ")
            print("]\n")
    

# printing the solution
current = "G213"
print(str(current[:-2]) + " was found with the cost of " + str(visited_dic.get("G2")))
print("\nRead from bottom to top to know the path")
print(str(current[:-2]))

# While loop to print path to the goal nodes
while current != start_node:
    listToStr = ''.join(str(elem) for elem in current)
    if listToStr == "S0":
            break
    current = tracking_dic[listToStr]
    print("↑")
    print(current[0])
```
